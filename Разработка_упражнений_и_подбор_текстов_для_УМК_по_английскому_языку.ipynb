```python
import re

# Функция для загрузки файла
def load_file(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        return f.read()

# Загрузка текста и списка глаголов
text = load_file('text.txt')
verbs_list = load_file('words.txt').splitlines()

# Создание регулярных выражений для каждого глагола, чтобы заменить их в тексте
patterns = [(re.compile(r'\b{}\b'.format(re.escape(verb)), re.IGNORECASE), verb) for verb in verbs_list]

# Заменяем глаголы на нижние подчеркивания
def replace_verbs(text, patterns):
    for pattern, verb in patterns:
        # Создаем подстановочный текст
        underscore = '_' * len(verb)
        # Замена с сохранением регистра
        def replace_match(match):
            # Регистронезависимая замена
            original = match.group()
            # Если оригинал с большой буквы, делаем подчеркивания с большой буквы
            if original[0].isupper():
                return underscore.capitalize()
            else:
                return underscore
        text = pattern.sub(replace_match, text)
    return text

# Получаем текст с замененными глаголами
exercise_text = replace_verbs(text, patterns)

# Оформление упражнения
print("Заполните пропуски, вставляя соответствующие глаголы:\n")
print(exercise_text)
print("\nСлова для вставки:\n")
for word in verbs_list:
    print(word)
```

2) Заменить время написанное словами на цифры: half past eight – 8:30, ten past three – 3:10, seven – 7, half past ten – 10:30, quarter to one – 12:45.

```python
import re

# Словарь чисел
numbers = {
    "one": 1,
    "two": 2,
    "three": 3,
    "four": 4,
    "five": 5,
    "six": 6,
    "seven": 7,
    "eight": 8,
    "nine": 9,
    "ten": 10,
    "eleven": 11,
    "twelve": 12,
    "thirteen": 13,
    "fourteen": 14,
    "fifteen": 15,
    "sixteen": 16,
    "seventeen": 17,
    "eighteen": 18,
    "nineteen": 19,
    "twenty": 20,
    "twenty one": 21,
    "twenty two": 22,
    "twenty three": 23,
    "twenty four": 24
}

# Функция для получения числа из текста
def words_to_number(word):
    word = word.lower()
    return numbers.get(word, None)

def parse_time_phrase(phrase):
    phrase = phrase.lower().strip()

    # Обработка "half past" -> xx:30
    match_half_past = re.match(r'half past (\w+)', phrase)
    if match_half_past:
        hour_word = match_half_past.group(1)
        hour = words_to_number(hour_word)
        if hour:
            return f"{hour}:30"
        else:
            return "Не распознан час"

    # Обработка "quarter past" -> xx:15
    match_quarter_past = re.match(r'quarter past (\w+)', phrase)
    if match_quarter_past:
        hour_word = match_quarter_past.group(1)
        hour = words_to_number(hour_word)
        if hour:
            return f"{hour}:15"
        else:
            return "Не распознан час"

    # Обработка "quarter to" -> следующий час : 45
    match_quarter_to = re.match(r'quarter to (\w+)', phrase)
    if match_quarter_to:
        hour_word = match_quarter_to.group(1)
        hour = words_to_number(hour_word)
        if hour:
            next_hour = hour + 1 if hour < 12 else 1
            return f"{next_hour}:45"
        else:
            return "Не распознан час"

    # Простое число (например, "seven")
    match_simple = re.match(r'(\w+)', phrase)
    if match_simple:
        hour_word = match_simple.group(1)
        hour = words_to_number(hour_word)
        if hour:
            return f"{hour}:00"
        else:
            return "Не распознан час"

    return "Не распознано"
# Примеры
examples = [
    "half past eight",
    "ten past three",
    "seven",
    "half past ten",
    "quarter to one",
    "quarter past four",
    "ten"
]

for example in examples:
    print(f"{example} -> {parse_time_phrase(example)}")

# Ввод фразы с клавиатуры
user_input = input("Введите фразу времени на английском: ")
result = parse_time_phrase(user_input)
print(f"Результат: {result}")
```

```python
import re

# Словарь для замены чисел
num_to_words = {
    '0': 'zero', '1': 'one', '2': 'two', '3': 'three', '4': 'four',
    '5': 'five', '6': 'six', '7': 'seven', '8': 'eight', '9': 'nine',
    '10': 'ten', '11': 'eleven', '12': 'twelve', '13': 'thirteen',
    '14': 'fourteen', '15': 'fifteen', '16': 'sixteen', '17': 'seventeen',
    '18': 'eighteen', '19': 'nineteen', '20': 'twenty', '21': 'twenty one',
    '22': 'twenty two', '23': 'twenty three', '24': 'twenty four',
    '25': 'twenty five', '26': 'twenty six', '27': 'twenty seven',
    '28': 'twenty eight', '29': 'twenty nine', '30': 'thirty'
}

# Открываем файл и читаем текст
with open("text.txt", "r", encoding="utf-8") as file:
    text = file.read()

# 1. Замена сокращений с помощью регулярных выражений
contractions = {
    r"\bI’m\b": "I am",
    r"\bname’s\b": "name is",
    r"\bI’ve\b": "I have",
    r"\bwe’ve\b": "we have",
    r"\bthey’re\b": "they are",
    r"\bthey’ve\b": "they have",
    r"\bcan’t\b": "cannot",
    r"\bthat’s\b": "that is",
    r"\bwhat’s\b": "what is",
}

for pattern, full_form in contractions.items():
    text = re.sub(pattern, full_form, text, flags=re.IGNORECASE)

# 2. Замена цифр в словах на слова
def replace_numbers(match):
    number_str = match.group()
    return num_to_words.get(number_str, number_str)

# Заменяем числа в диапазоне от 0 до 30
pattern_numbers = r"\b(" + "|".join(re.escape(k) for k in num_to_words.keys()) + r")\b"
text = re.sub(pattern_numbers, replace_numbers, text)

# 3. Замена времени в формате "half past eight", "ten past three" и т.д.
# Обрабатываем конкретные случаи из текста
time_patterns = {
    r"\bhalf past (\w+)\b": lambda m: f"{convert_hour(m.group(1))}:30",
    r"\bten past (\w+)\b": lambda m: f"{convert_hour(m.group(1))}:10",
    r"\bquarter to (\w+)\b": lambda m: f"{convert_hour(m.group(1))}:45",
    r"\bquarter past (\w+)\b": lambda m: f"{convert_hour(m.group(1))}:15",
    r"\b(\w+) o’clock\b": lambda m: f"{convert_hour(m.group(1))}:00",
}

def convert_hour(word):
    # преобразование слова в число
    hour_map = {
        "one": 1, "two": 2, "three": 3, "four": 4, "five": 5,
        "six": 6, "seven": 7, "eight": 8, "nine": 9, "ten": 10,
        "eleven": 11, "twelve": 12
    }
    return str(hour_map.get(word.lower(), word))

# Обработка случаев времени
for pattern, func in time_patterns.items():
    text = re.sub(pattern, lambda m: func(m), text, flags=re.IGNORECASE)

# 4. Замена слов в формате "... at ..." для времени
# Любое упоминание времени наподобие "at half past eight" и т.п.
# Необходимо расширить при необходимости.

# Итог: сохраняем или выводим
print("Обработанный текст:")
print(text)

# Можно сохранить в файл
with open("text.txt", "w", encoding="utf-8") as file:
    file.write(text)
```

```python
import os
import re
import matplotlib.pyplot as plt

# Функция для чтения вокабуляра
def load_vocabulary(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        line = f.read().strip()
        line = line.rstrip('.')
        vocabulary = [word.strip().lower() for word in line.split(',')]
    return vocabulary

# Функция для подсчета совпадений в тексте
def count_vocab_matches(text, vocab_set):
    words = re.findall(r'\b\w+\b', text.lower())
    count = sum(1 for word in words if word in vocab_set)
    return count

# 1. Загружаем вокабуляр
vocab = load_vocabulary('vocabulary.txt')
vocab_set = set(vocab)

results = []
filenames = []

# 2. Анализируем каждый текстовый файл
for i in range(1, 11):
    filename = f'text{i}.txt'
    with open(filename, 'r', encoding='utf-8') as f:
        text = f.read()
    total_words = len(re.findall(r'\b\w+\b', text))
    match_count = count_vocab_matches(text, vocab_set)
    percentage = (match_count / total_words * 100) if total_words > 0 else 0
    results.append(percentage)
    filenames.append(filename)

# 3. Выводим статистику
for filename, perc in zip(filenames, results):
    print(f'{filename}: {perc:.2f}% совпадений лексики из вокабуляра')

# 4. Строим диаграмму
plt.figure(figsize=(10, 6))
plt.bar(filenames, results, color='skyblue')
plt.xlabel('Файлы')
plt.ylabel('Процент совпадений (%)')
plt.title('Статистика совпадений лексики из вокабуляра по файлам')
plt.xticks(rotation=45)
plt.ylim(0, 40)

# 5. Выводим самый подходящий текст
best_idx = results.index(max(results))
print(f"\nЛучший текст: {filenames[best_idx]} с {results[best_idx]:.2f}% совпадений.")

# Показываем график
plt.tight_layout()
plt.show()
```

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import os
from wordcloud import WordCloud

# Ваша существующая функция
def get_top_words(text, top_n=10):
    words = re.findall(r'\b\w+\b', text.lower())
    # Отфильтруем слова длиной > 4 символов
    filtered_words = [w for w in words if len(w) > 4]
    counts = Counter(filtered_words).most_common(top_n)
    return [word for word, count in counts]

# Чтение файлов
texts = []
for i in range(1, 11):
    filename = f"text{i}.txt"
    if os.path.exists(filename):
        with open(filename, 'r', encoding='utf-8') as file:
            texts.append(file.read())
    else:
        print(f"Предупреждение: Файл {filename} не найден.")

data = {}
all_selected_words = []
# Обработка каждого текста
for i, text in enumerate(texts, 1):
    top = get_top_words(text)
    all_selected_words.extend(top)
    full_list = top + [None] * (10 - len(top))
    data[f"Текст {i}"] = full_list

if texts:
    df = pd.DataFrame(data, index=range(1, 11))
    print("Таблица самых частотных слов (длина > 4 символа):")
    print(df)

    # Статистика для облака
    if all_selected_words:
        word_counts = Counter(all_selected_words)
        common = word_counts.most_common(10)
        words_labels = [item[0] for item in common]
        counts_values = [item[1] for item in common]

        plt.figure(figsize=(12, 6))
        plt.bar(words_labels, counts_values, color='skyblue')
        plt.title('Топ-10 слов (>4 символов) среди всех текстов')
        plt.xlabel('Слова')
        plt.ylabel('Частота появления')
        plt.xticks(rotation=45)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

        # Создаем словарь для WordCloud из топ-слов
        top_words_freq = dict(common)  # {"слово": частота, ...}

        # Генерируем облако
        wc = WordCloud(
            width=800,
            height=600,
            colormap='viridis',
            random_state=42
        ).generate_from_frequencies(top_words_freq)

        # Отобразим облако
        plt.figure(figsize=(10, 8))
        plt.imshow(wc, interpolation='bilinear')
        plt.axis('off')
        plt.title('Облако слов из топовых слов (>4 символов)')
        plt.show()

else:
    print("Ошибка: ни один файл не был прочитан.")
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import random

# Ввод 20 слов вручную
words = []
print("Введите 20 слов:")
for i in range(20):
    word = input(f"Слово {i+1}: ")
    words.append(word)

# Создаем словарь с частотами, присваивая каждому слово случайную частоту
text = {}
for word in words:
    frequency = random.randint(10, 100)
    text[word] = frequency

# Функция для случайного окраса слов
def random_color_func(word, **kwargs):
    return "rgb({}, {}, {})".format(
        random.randint(0, 255),
        random.randint(0, 255),
        random.randint(0, 255)
    )

# Создаем объект WordCloud с белым фоном
wc = WordCloud(
    width=800,
    height=600,
    background_color='white',  # задаем белый фон
    random_state=42,
    colormap=None,
    color_func=random_color_func,
    prefer_horizontal=0.5,
    min_font_size=10,
    max_font_size=100,
)

# Генерируем облако
wc.generate_from_frequencies(text)

# Отрисовываем
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
